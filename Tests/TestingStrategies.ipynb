## Тестирование тотал меньше
Условие: Если тотал меньше прематч коэфф стоит между 1.5 и 1.8 и количество голов меньше 2.5, то победа, в противном случае получаем поражение

```python
bank = 10000
percent = 0.02
for i in range(0,len(data)):
    if data['B365<2.5'][i] > 1.5 and data['B365<2.5'][i] < 1.8 and data['FTG'][i] < 3:
        bank = (bank + (bank * data['B365<2.5'][i] * percent)) - bank * percent
    elif data['B365<2.5'][i] > 1.5 and data['B365<2.5'][i] < 1.8 and data['FTG'][i] > 2:
        bank = bank - bank * percent
bank    
```

## Тестирование тотал больше
Условие: Если тотал больше прематч коэфф стоит между 1.5 и 1.8 и количество голов больше 2.5, то победа, в противном случае получаем поражение

```python
bank = 10000
percent = 0.02
for i in range(0,len(data)):
    if data['B365>2.5'][i] > 1.5 and data['B365>2.5'][i] < 1.8 and data['FTG'][i] > 2:
        bank = (bank + (bank * data['B365>2.5'][i] * percent)) - bank * percent
    elif data['B365>2.5'][i] > 1.5 and data['B365>2.5'][i] < 1.8 and data['FTG'][i] < 3:
        bank = bank - bank * percent
bank
```

## Тестирование победа дома
Условие: Если коэфф победа дома больше 1.5 и меньше 1.8, и домашняя команда победила - победа, в противном случае при этих коэфф поражение

```python
bank = 10000
percent = 0.02
for i in range(0,len(data)):
    if data['B365H'][i] > 1.5 and data['B365H'][i] < 1.8 and data['FTR'][i] == 'H':
        bank = (bank + (bank * data['B365H'][i] * percent)) - bank * percent
    elif data['B365H'][i] > 1.5 and data['B365H'][i] < 1.8 and data['FTR'][i] != 'H':
        bank = bank - bank * percent
bank
```
